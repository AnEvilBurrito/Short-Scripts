```python
import yfinance as yf

```

```python
# download vix and sp500 data 

start = "2000-01-01"
end = "2022-12-31"

VIX_data = yf.download("^VIX", start=start, end=end)
SP500_data = yf.download("^GSPC", start=start, end=end)
```

```python
VIX_data.head()
```

```python
SP500_data.head()
```

```python
SP500_data_close = SP500_data['Close']
VIX_data_close = VIX_data['Close']
```

```python
# log change of sp500 

import numpy as np

SP500_data_close_log = np.log(SP500_data_close)
```

```python
# plot sp500 log change

import matplotlib.pyplot as plt

plt.plot(SP500_data_close_log)
plt.show()
```

```python
import matplotlib.pyplot as plt

# two subplots, the axes array is 1-d

f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(SP500_data_close_log)
axarr[0].set_title('SP500 (Log price)')
axarr[0].grid(True)
axarr[1].plot(VIX_data_close, 'r', alpha=0.5)
axarr[1].set_title('VIX')
axarr[1].grid(True)
plt.show()
```
